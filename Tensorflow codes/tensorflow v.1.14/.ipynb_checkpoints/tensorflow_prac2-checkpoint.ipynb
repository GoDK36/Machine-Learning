```python
## Linear Regression ##
import tensorflow as tf
```

```python
# X and Y data

x_train = [1, 2, 3]
y_train = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name='weight')  ## tensorflow가 사용하는 변경이 가능한 값(훈련이 가능한 값이다)
b = tf.Variable(tf.random_normal([1]), name='bias')  ## 1차원이라는 뜻

# Hypothesis
hypothesis = x_train * W + b
```

```python
# Cost(loss) function

cost = tf.reduce_mean(tf.square(hypothesis - y_train))  # resuce_mean은 평균 내주는것
```

```python
# Minimize

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)  # 경사하강법
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())  # Variable을 설정하면 실행 전 반드시 이 코드가 필요함

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
## Placeholder  :  직접 값을 주지 않고 사용
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b  # tf.add(a, b)와 동일

print(sess.run(adder_node, feed_dict={a: 3, b: 4.5}))
print(sess.run(adder_node, feed_dict={a: [1, 3], b: [2, 4]}))
```

```python
X = tf.placeholder(tf.float32, shape=[None]) ## 아무값이나 들어올수있다. shape=[None] => 아무 형태나 입력 가능
Y = tf.placeholder(tf.float32, shape=[None])

hypothesis = X * W + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())  # Variable을 설정하면 실행 전 반드시 이 코드가 필요함

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run(
                [cost, W, b, train],  # 리스트 형싟으로 한 번에 훈련 가능
                feed_dict={X: [1, 2, 3, 4, 5], Y: [2.1, 3.1, 4.1, 5.1, 6.1]}
                )  # sess.run에서 feed_dict로 값 주기 가능
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```

```python
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```

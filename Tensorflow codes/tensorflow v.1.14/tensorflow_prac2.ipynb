```python
## Linear Regression ##
import tensorflow as tf
```

```python
# X and Y data

x_train = [1, 2, 3]
y_train = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name='weight')  ## tensorflow가 사용하는 변경이 가능한 값(훈련이 가능한 값이다)
b = tf.Variable(tf.random_normal([1]), name='bias')  ## 1차원이라는 뜻

# Hypothesis
hypothesis = x_train * W + b
```

```python
# Cost(loss) function

cost = tf.reduce_mean(tf.square(hypothesis - y_train))  # resuce_mean은 평균 내주는것
```

```python
# Minimize

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)  # 경사하강법
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())  # Variable을 설정하면 실행 전 반드시 이 코드가 필요함

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")

X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

# Our hypothesis is X * W + b
hypothesis = X * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# optimizer
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())  # Variable을 설정하면 실행 전 반드시 이 코드가 필요함

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run(
                [cost, W, b, train],  
                feed_dict={X: [1, 2, 3, 4, 5], Y: [2.1, 3.1, 4.1, 5.1, 6.1]}
                )  
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
        
for step in range(2001):
        _, cost_val, W_val, b_val = sess.run(
            [train, cost, W, b],  # 리스트 형싟으로 한 번에 훈련 가능 
            feed_dict={X: [1, 2, 3], Y: [1, 2, 3]} # sess.run에서 feed_dict로 값 주기 가능
        )
        if step % 20 == 0:
            print(step, cost_val, W_val, b_val)
```

```python
# training data
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]

# 모델 파라미터
W = tf.Variable([0.3], tf.float32)
b = tf.Variable([-0.3], tf.float32)

x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)

hypothesis = x * W + b

cost = tf.reduce_sum(tf.square(hypothesis - y))

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

# 모델 훈련하기
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(1000):
    sess.run(train, {x: x_train, y: y_train})

# 정확도 계산하기
W_val, b_val, cost_val = sess.run([W, b, cost], feed_dict={x: x_train, y: y_train})
print(f"W: {W_val} b: {b_val} cost: {cost_val}")
```
